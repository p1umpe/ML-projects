#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_letters(sttr):
    letters = ['a', 'e', 'i', 'o', 'u']
    count = 0
    for i in sttr.lower():
        if i in letters:
            count +=1
    return count

def test_count_letter():
    assert count_letters("aeiou") == 5
    print("Тест на все гласные пройден")

    assert count_letters("bcdfg") == 0
    print("Тест на все согласные пройден")

    assert count_letters("") == 0
    print("Тест пустая строка пройден")

    assert count_letters("AEIOU") == 5
    print("Тест на только заглавные пройден")

    assert count_letters("f g a k o") == 2
    print("Тест если есть пробелы пройден")

test_count_letter()

if __name__ == "__main__":
    text = input()
    result = count_letters(text)
    print(result)
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def search_letters(sttr):
    seen_letters = set()
    for i in sttr:
        if i in seen_letters:
            return False
        else:
            seen_letters.add(i)
    return True

def test_search_letters():
    assert search_letters("abcdef") == True
    print("Тест все уникальны пройден")

    assert search_letters("hello") == False
    print("Тест есть повторы пройден")

    assert search_letters("") == True
    print("Тест пустая строка пройден")

    assert search_letters("a") == True
    print("Тест 1 символ пройден")

    assert search_letters("aA") == True
    print("Тест если разный регистр пройден")

test_search_letters()

if __name__ == "__main__":
    text = input()
    result = search_letters(text)
    print(result)
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits_bin(n):
    return bin(n).count('1')

def test_count_bits():
    assert count_bits_bin(0) == 0
    print("Тест 1 пройден")
    
    assert count_bits_bin(1) == 1
    print("Тест 2 пройден")
    
    assert count_bits_bin(2) == 1
    print("Тест 3 пройден")
    
    assert count_bits_bin(3) == 2
    print("Тест 4 пройден")
    
    assert count_bits_bin(5) == 2
    print("Тест 5 пройден")

    assert count_bits_bin(1023) == 10
    print("Тест 6 пройден")

    assert count_bits_bin(255) == 8
    print("Тест 7 пройден")

if __name__ == "__main__":
    test_count_bits()

    numbers = int(input())
    if numbers < 0:
        print("Число должно быть положительным")
    else:
        result = count_bits_bin(numbers)
        print(result)

```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    if n < 10:
        return 0
    
    steps = 0
    current = n

    while current >= 10:
        result = 1
        for i in str(current):
            result *= int(i)
        current = result
        steps +=1
    return steps

def test_magic():
    assert magic(4) == 0
    print("Тест 1 пройден")
    
    assert magic(39) == 3
    print("Тест 2 пройден")
    
    assert magic(999) == 4
    print("Тест 3 пройден")
    
    assert magic(10) == 1
    print("Тест 4 пройден")
    
    assert magic(77) == 4
    print("Тест 5 пройден")
    
    assert magic(0) == 0
    print("Тест 6 пройден")
    
    assert magic(6788) == 6
    print("Тест 7 пройден")

if __name__ == "__main__":
    test_magic()

    number = int(input())
    if number < 0:
        print("Число должно быть положительным")
    else:
        result = magic(number)
        print(f"Для числа {number} потребовалось {result} шагов")
        
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    total_error = 0
    for i in range(len(pred)):
        error = pred[i] - true[i]
        total_error += error ** 2
    
    return total_error / len(pred)

def test_mse():
    assert mse([1, 2, 3], [1, 2, 3]) == 0
    print("Тест идентичные векторы")
    
    assert mse([2, 3, 4], [1, 2, 3]) == 1
    print("Тест постоянное отклонение")
    
    result = mse([3, 5, 7], [1, 2, 3])
    expected = ((2**2 + 3**2 + 4**2) / 3)
    assert abs(result - expected) < 1e-10
    print("Тест разные отклонения")
    
    assert mse([-1, -2, -3], [-2, -3, -4]) == 1
    print("Тест отрицательные числа")
    
    assert mse([5], [3]) == 4
    print("Тест один элемент")

    result = mse([1.5, 2.5], [1.0, 2.0])
    expected = ((0.5**2 + 0.5**2) / 2)
    assert abs(result - 0.25) < 1e-10
    print("Тест дробные числа")

if __name__ == "__main__":
    test_mse()
    pred_input = input()
    true_input = input()

    pred = [float(x) for x in pred_input.split()]
    true = [float(x) for x in true_input.split()]

    result = mse(pred, true)
    print(result)
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def multiplication(n):
    if n <= 1:
        return f"({n})"

    factors = {}
    div = 2

    while div * div <= n:
        while n % div == 0:
            factors[div] = factors.get(div, 0) + 1
            n //= div
        div += 1

    if n > 1:
        factors[n] = factors.get(n, 0) + 1

    result = []
    for factor in sorted(factors.keys()):
        power = factors[factor]
        if power == 1:
            result.append(f"({factor})")
        else:
            result.append(f"({factor}**{power})")
    
    return "".join(result)

def test_multiplication():
    assert multiplication(13) == "(13)"
    print("Тест простое число пройден")
    
    assert multiplication(86240) == "(2**5)(5)(7**2)(11)"
    print("Тест условие пройден")
    
    assert multiplication(8) == "(2**3)"
    print("Тест степень 2 пройден")
    
    assert multiplication(1) == "(1)"
    print("Тест число 1 пройден")
    
    assert multiplication(30) == "(2)(3)(5)"
    print("Тест простые множители пройден")
    
    assert multiplication(100) == "(2**2)(5**2)"
    print("Тест пройден: 100")

if __name__ == "__main__":
    test_multiplication()
    number = int(input())
    if number < 0:
        print("Число должно быть положительным")
    else:
        result = multiplication(number)
        print(result)



```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    if number <= 0:
        return "It is impossible"
    
    k = 1
    while True:
        total = k * (k + 1) * (2 * k + 1) // 6
        
        if total == number:
            return k
        elif total > number:
            return "It is impossible"
        k += 1

def test_pyramid():
    assert pyramid(1) == 1
    print("Тест 1 пройден")
    
    assert pyramid(5) == 2
    print("Тест 2 пройден")
    
    assert pyramid(14) == 3
    print("Тест 3 пройден")
    
    assert pyramid(30) == 4
    print("Тест 4 пройден")
    
    assert pyramid(55) == 5
    print("Тест 5 пройден")
    
    assert pyramid(2) == "It is impossible"
    print("Тест 6 пройден")
    
    assert pyramid(0) == "It is impossible"
    print("Тест 7 пройден")
    
    assert pyramid(-5) == "It is impossible"
    print("Тест 8 пройден")

if __name__ == "__main__":
    test_pyramid()

    cubes = int(input())
    result = pyramid(cubes)
        
    if isinstance(result, int):
        print(f"Можно построить пирамиду высотой {result}")
    else:
        print("It is impossible")
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(number):
    digits = [int(d) for d in str(number)]
    n = len(digits)
    
    if n <= 2:
        return False
    if n % 2 == 1:
        mid = n // 2
        left_sum = sum(digits[:mid])
        right_sum = sum(digits[mid + 1:])
        return left_sum == right_sum
    else:
        mid = n // 2
        left_sum = sum(digits[:mid - 1])
        right_sum = sum(digits[mid + 1:])
        return left_sum == right_sum

def test_is_balanced():
    assert is_balanced(12321) == True
    print("Тест 1 пройден")
    
    assert is_balanced(123321) == True
    print("Тест 2 пройден")
    
    assert is_balanced(12345) == False
    print("Тест 3 пройден")
    
    assert is_balanced(123456) == False
    print("Тест 4 пройден")
    
    assert is_balanced(121) == True
    print("Тест 5 пройден")
    
    assert is_balanced(123) == False
    print("Тест 6 пройден")

    assert is_balanced(1221) == True
    print("Тест 7 пройден")

    assert is_balanced(1234) == False
    print("Тест 8 пройден")

    assert is_balanced(12) == False
    print("Тест 9 пройден")

    assert is_balanced(1) == False
    print("Тест 10 пройден")

    assert is_balanced(123404321) == True
    print("Тест 11 пройден")

    assert is_balanced(123456789) == False
    print("Тест 12 пройден")

    assert is_balanced(1001) == True
    print("Тест 13 пройден")

    assert is_balanced(11) == False
    print("Тест 14 пройден")

if __name__ == "__main__":
    test_is_balanced()

    number = int(input())
    if number < 0:
        print("Число должно быть положительным")
    else:
        result = is_balanced(number)
        if result:
            print(f"Число {number} сбалансировано")
        else:
            print(f"Число {number} не сбалансировано")

```
